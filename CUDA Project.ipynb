**CPU Runtime in Grayscaling an image**

```python
# Menggunakan library NumPy (Numerical Python) untuk melakukan grayscale pada gambar
# Kalkulasi CPU Time

import matplotlib.pyplot as plt

start = time.time()
def RGB_to_Gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])
image = mpimage.imread('machupicchu.jpg')
gray = RGB_to_Gray(image)
end = time.time()
plt.imshow(gray, cmap=plt.get_cmap('gray'))
plt.show()

print("CPU Time: ", end - start)
```

**GPU Runtime in Grayscaling an image**

```python
# Menggunakan library CuPy (Cuda Python) untuk melakukan grayscale pada gambar dengan weight
# Kalkulasi GPU Time

import matplotlib.pyplot as plt

image = cupy.array(mpimage.imread('machupicchu.jpg'))
gpu_weight = cupy.array([0.299, 0.587, 0.114])
start = time.time()

def RGB_to_Gray(rgb):
    return cupy.dot(image[...,:3], gpu_weight)

gray = RGB_to_Gray(image)
end = time.time()
plt.imshow(gray.get(), cmap=plt.get_cmap('gray'))
plt.show()
print("GPU time: ", end - start)
```

#**Conclusion**
**GPU > CPU**

In this CUDA Project, we conclude that in Grayscaling an image, it will fasten the Grayscaling process if we use CUDA as we can see that the GPU Runtime is faster than the CPU Runtime.
